```python
# where 过滤
ndaifa=np.array(['a','b','c','d','e'])
ndaifb=np.array(['A','B','C','D','E'])
ndaif=np.array([True,False,False,True,True])
nadif_result=np.where(ndaif,ndaifa,ndaifb)
nadif_result

```

```python
# 元素级数组函数
import numpy  

arr2=numpy.random.randint(0,100,16)
numpy.square(arr2)

# 是数组对象中的方法，用于改变数组的形状

print("原始数据",arr2,"\nreshape数据",arr2.reshape(4,4))
```

```python
import numpy as np

arn2=np.random.randn(5,4)
print(arn2)
# 算术平均值
print("算术平均值",arn2.mean())

# 定轴统计
print("0维上的和",arn2.sum(axis=0))

arn3=np.random.randint(0,10,10)

print("多个随机整数生成",arn3)
print("累计和",arn3.cumsum())
print("累计积",arn3.cumprod())
```

```python
# 排序
import numpy  

arn4=numpy.random.randint(0,10,10)
print("原数据：",arn4)
arn4.sort()
print("排序后：",arn4)
```

```python
# 逻辑函数
# 排序
import numpy  

arn4=numpy.random.randint(0,10,10)
arn4_1=numpy.random.randint(0,10,4)
print("原数据：",arn4)
print("数组去重：",set(arn4.copy()))
print("数组去重，并排序：",numpy.unique(arn4))
print("非：",arn4,arn4_1,numpy.setdiff1d(arn4,arn4_1))
print("亦或：",arn4,arn4_1,numpy.setxor1d(arn4,arn4_1))


```
