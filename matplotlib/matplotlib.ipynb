```python
# Figure ，axes对象 

import matplotlib.pyplot as plt
import numpy as np
 
fig=plt.figure()

add_subplot1=fig.add_subplot(2,2,1)
# 利用axessubplot对象绘制
add_subplot1.hist(np.random.randn(10),bins=20,color='k',alpha=0.3)

# 绘制点
add_subplot2=fig.add_subplot(2,2,2)
add_subplot2.scatter(np.arange(30),np.arange(30)+3*np.random.randn(30))

# 绘制虚线
add_subplot3=fig.add_subplot(2,2,3)
a=np.array([0,1,3,1,4,5])
plt.plot(a,'k--',color='#23CEF2')
plt.plot(a.cumsum(),'k-')

# 设置刻度范围
plt.xlim([0,10])

# 设置刻度的标签
add_subplot3.set_xlabel('sss')
```

```python
# 随机漫步

import matplotlib.pyplot as plt
import numpy as np

fig=plt.figure();
ax=fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum(),'k.',label='one')

# legend标识图表元素 
# loc设置图例位置
ax.legend(loc='best')
```

```python
# 保存图表 到 文件
# 格式，分辨率，空白(bbox_inches 剪除空白 ),颜色
import matplotlib.pyplot as plt
import numpy as np

fig=plt.figure();
ax=fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum(),'k.',label='one')

# legend标识图表元素 
# loc设置图例位置
ax.legend(loc='best')
plt.savefig('f.svg',dpi=300,bbox_inches='tight',edgecolor='w')
```
